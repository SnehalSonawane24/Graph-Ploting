```python
#line graphs
import matplotlib.pyplot as plt
a =[1,2,3]
b =[20,50,80]
plt.plot(a,b)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
```

```python
import matplotlib.pyplot as plt
a = [1,2,3,4]
b = [120,130,135,140]
plt.bar(a,b)
plt.title('roll num vs marks')
plt.xlabel('roll num')
plt.ylabel('marks')
plt.show()
```

```python
#Barchart Using Matplot
import matplotlib.pyplot as plt
a = [1,2,3,4]
b = [86,63,70,89]
plt.bar(a,b)
plt.xlabel('Roll_num')
plt.ylabel('Marks_Obtain')
plt.show()
```

```python
#Scatter Plot : Each value in Dataset is reperensted by dots
import matplotlib.pyplot as plt
a = [1,27,89,45,67,35]
b = [13,97,57,43,39,21]
plt.scatter(a,b)
plt.show()
```

```python
# Histogram : Graph that shows frequency distribution
import matplotlib.pyplot as plt
import numpy as np
x = np.random.normal(170,10,250)
plt.hist(x)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.random.normal(60,120,180)
plt.hist(x)
plt.show()
```

```python
#BoxPlot: whisker plot is creted to display the summary of set of data values having properties.
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(10)
x = np.random.normal(100,20,200)
plt.boxplot(x)
plt.show()
```

```python
#piechart: Circular Statistical plot that can display only one series of data.
cars = ['AUDI', 'BMW', 'NISSAN','TESLA', 'HYUNDAI', 'HONDA'] 
data = [20, 15, 15, 14, 16, 20]
#Creating the pie chart
plt.pie(data, labels = cars)
#Adding the aesthetics 
plt.title('PIE CHART') 
#Show the plot
plt.show()
```
